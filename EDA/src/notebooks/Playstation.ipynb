```python
from bs4 import BeautifulSoup as bs
import requests
import pandas as pd
import numpy as np
```

```python
generos = ['Role-Playing','Action','Adventure','Action-Adventure','Misc','Platform','Puzzle','Racing','Shooter','Simulation','Sports','Strategy']

lista_de_dataframes = []

for genero in generos:
        url = f'https://www.vgchartz.com/games/games.php?name=&keyword=&console=PS&region=All&developer=&publisher=&goty_year=&genre={genero}&boxart=Both&banner=Both&ownership=Both&showmultiplat=No&results=200&order=Sales&showtotalsales=0&showtotalsales=1&showpublisher=0&showvgchartzscore=0&shownasales=0&shownasales=1&showdeveloper=0&showcriticscore=0&showpalsales=0&showpalsales=1&showreleasedate=0&showreleasedate=1&showuserscore=0&showjapansales=0&showjapansales=1&showlastupdate=0&showothersales=0&showothersales=1&showshipped=0'
      
        df_temporal = pd.read_html(url)[-2]
        
        # Corregimos las columnas
        df_temporal.columns = df_temporal.columns.get_level_values(2)
        
        # Añadimos una columna para saber a qué género pertenece cada juego
        df_temporal['Genre'] = genero
        
        #  Añadimos el DataFrame limpio a nuestra lista usando .append()
        lista_de_dataframes.append(df_temporal)
        
ps1_completo = pd.concat(lista_de_dataframes, ignore_index=True)
ps1_completo
ps1_completo=ps1_completo[['Game.1','Total Sales','NA Sales','PAL Sales','Japan Sales','Other Sales','Release Date','Genre']]
```

```python
nuevos_nombres = [
    'Nombre',
    'Ventas Totales',
    'Ventas Norte América',
    'Ventas Europa',
    'Ventas Japón',
    'Ventas Otros',
    'Fecha Salida',
    'Género'
]
ps1_completo.columns = nuevos_nombres
```

```python
columnas_ventas = [
    'Ventas Totales',
    'Ventas Norte América',
    'Ventas Europa',
    'Ventas Japón',
    'Ventas Otros'
]
ps1_completo[columnas_ventas] = ps1_completo[columnas_ventas].apply(lambda x: x.str.replace('m', ''))
ps1_completo[columnas_ventas] = ps1_completo[columnas_ventas].apply(pd.to_numeric, errors='coerce')
```

```python
ps1_completo[['Ventas Totales',
    'Ventas Norte América',
    'Ventas Europa',
    'Ventas Japón',
    'Ventas Otros']].dropna(axis=0,how="all",inplace=True)
```
